# Maternal Health Risk Prediction

## Interactive AI Chatbot

#### Import Required Libraries

```python
import numpy as np
import pandas as pd
import pickle
import warnings

warnings.filterwarnings("ignore")

print("🤖 Maternal Health Risk Chatbot is starting...")
```

#### Load Best ANN Model

```python
print("="*70)
print("Loading Trained ANN Model")
print("="*70)

MODEL_PATH = "../Model/ann.pkl"

try:

    with open(MODEL_PATH, "rb") as file:
        model = pickle.load(file)

    print("✅ ANN Model Loaded Successfully.")

except Exception as e:

    print("❌ Error Loading Model")
    print(e)

print("="*70)
```

#### Feature Engineering Function

```python
print("="*70)
print("Creating Feature Engineering Function")
print("="*70)

def preprocess_input(df):

    df = df.copy()

    eps = 1e-3

    # Pulse Pressure
    if {"SystolicBP", "DiastolicBP"}.issubset(df.columns):

        df["PulsePressure"] = (
            df["SystolicBP"] -
            df["DiastolicBP"]
        )

        df["BPRatio"] = (
            df["SystolicBP"] /
            (df["DiastolicBP"] + eps)
        )

    # Blood Sugar × Heart Rate
    if {"BS", "HeartRate"}.issubset(df.columns):

        df["BS_HR_Interaction"] = (
            df["BS"] *
            df["HeartRate"]
        )

    # Age × Systolic BP
    if {"Age", "SystolicBP"}.issubset(df.columns):

        df["Age_SystolicBP"] = (
            df["Age"] *
            df["SystolicBP"]
        )

    return df

print("✅ Feature Engineering Function Ready.")
print("="*70)
```

#### Prediction Function

```python
print("="*70)
print("Creating Prediction Function")
print("="*70)

def predict_risk(input_data):

    # Convert dictionary to DataFrame
    df = pd.DataFrame([input_data])

    # Apply Feature Engineering
    df_processed = preprocess_input(df)

    # Predict Risk Level
    prediction = model.predict(df_processed)[0]

    return prediction

print("✅ Prediction Function Ready.")
print("="*70)
```

#### Risk Recommendation Function

```python
print("="*70)
print("Creating Recommendation Function")
print("="*70)

def get_recommendation(risk_level):

    risk = str(risk_level).lower()

    if risk == "low risk":

        return [
            "Maintain a healthy diet.",
            "Attend regular prenatal checkups.",
            "Stay physically active as advised by your doctor.",
            "Drink adequate water.",
            "Get sufficient rest."
        ]

    elif risk == "mid risk":

        return [
            "Consult your gynecologist regularly.",
            "Monitor blood pressure and blood sugar.",
            "Follow a balanced diet.",
            "Avoid stress and heavy physical work.",
            "Take prescribed medications on time."
        ]

    elif risk == "high risk":

        return [
            "Seek immediate medical attention.",
            "Monitor blood pressure frequently.",
            "Strictly follow your doctor's advice.",
            "Take medications without missing doses.",
            "Visit the hospital immediately if symptoms worsen."
        ]

    else:

        return [
            "Consult a healthcare professional."
        ]

print("✅ Recommendation Function Ready.")
print("="*70)
```

#### Interactive AI Chatbot

```python
print("="*70)
print("🤖 Maternal Health Risk Chatbot")
print("="*70)

print("\nType 'exit' anytime to quit.\n")

while True:

    try:

        # -------------------------------
        # Age
        # -------------------------------
        age = input("👩 Enter Age (in years): ")

        if age.lower() == "exit":
            break

        age = float(age)

        # -------------------------------
        # Blood Pressure
        # -------------------------------
        systolic = float(
            input("🩺 Enter Systolic BP (e.g., 120): ")
        )

        diastolic = float(
            input("🩺 Enter Diastolic BP (e.g., 80): ")
        )

        # -------------------------------
        # Blood Sugar
        # -------------------------------
        bs = float(
            input("🧪 Enter Blood Sugar (BS) value (e.g., 6.5): ")
        )

        # -------------------------------
        # Body Temperature
        # -------------------------------
        body_temp = float(
            input("🌡️ Enter Body Temperature (e.g., 98.6): ")
        )

        # -------------------------------
        # Heart Rate
        # -------------------------------
        heart_rate = float(
            input("❤️ Enter Heart Rate (e.g., 90): ")
        )

        patient = {

            "Age": age,
            "SystolicBP": systolic,
            "DiastolicBP": diastolic,
            "BS": bs,
            "BodyTemp": body_temp,
            "HeartRate": heart_rate

        }

        prediction = predict_risk(patient)

        print("\n" + "="*70)
        print("📋 PATIENT DETAILS")
        print("="*70)

        for key, value in patient.items():
            print(f"{key:<15}: {value}")

        print("-"*70)

        print(f"🩺 Predicted Risk Level : {prediction}")

        print("\n💡 Medical Recommendations")

        recommendations = get_recommendation(prediction)

        for i, tip in enumerate(recommendations, start=1):

            print(f"{i}. {tip}")

        print("="*70)

        again = input(
            "\n🔄 Do you want another prediction? (Y/N): "
        )

        if again.lower() != "y":
            break

        print()

    except ValueError:

        print("\n❌ Invalid Input.")
        print("Please enter numeric values only.\n")

    except Exception as e:

        print("\n⚠️ Something went wrong.")
        print(e)
        print()
```

#### Exit Message

```python
print("\n" + "="*70)

print("👋 Chatbot session ended.")

print("Stay Safe & Stay Healthy!")

print("="*70)
```
