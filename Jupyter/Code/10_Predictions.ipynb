# Maternal Health Risk Prediction

## Prediction Using Best ANN Model

#### Import Required Libraries

```python
import numpy as np
import pandas as pd
import pickle
import warnings

warnings.filterwarnings("ignore")

print("Starting Prediction Script...")
```

#### Load ANN Model

```python
print("="*70)
print("Loading Trained ANN Model")
print("="*70)

MODEL_PATH = "../Model/ann.pkl"

with open(MODEL_PATH, "rb") as file:
    model = pickle.load(file)

print("ANN Model Loaded Successfully.")
```

#### Feature Engineering

```python
print("="*70)
print("Feature Engineering Function")
print("="*70)

def preprocess_input(df):

    df = df.copy()

    eps = 1e-3

    if {"SystolicBP","DiastolicBP"}.issubset(df.columns):

        df["PulsePressure"] = (
            df["SystolicBP"] -
            df["DiastolicBP"]
        )

        df["BPRatio"] = (
            df["SystolicBP"] /
            (df["DiastolicBP"] + eps)
        )

    if {"BS","HeartRate"}.issubset(df.columns):

        df["BS_HR_Interaction"] = (
            df["BS"] *
            df["HeartRate"]
        )

    if {"Age","SystolicBP"}.issubset(df.columns):

        df["Age_SystolicBP"] = (
            df["Age"] *
            df["SystolicBP"]
        )

    return df

print("Feature Engineering Ready.")
```

#### Prediction Function

```python
print("="*70)
print("Prediction Function")
print("="*70)

def predict_risk(input_data):

    df = pd.DataFrame([input_data])

    df = preprocess_input(df)

    prediction = model.predict(df)[0]

    return prediction

print("Prediction Function Created.")
```

#### Sample Input

```python
print("="*70)
print("Sample Patient Details")
print("="*70)

sample_input = {

    "Age":28,

    "SystolicBP":130,

    "DiastolicBP":85,

    "BS":7.2,

    "BodyTemp":98.5,

    "HeartRate":92

}

sample_df = pd.DataFrame([sample_input])

display(sample_df)
```

#### Prediction

```python
print("="*70)
print("Running Prediction")
print("="*70)

prediction = predict_risk(sample_input)

print("Predicted Maternal Health Risk Level :", prediction)
```

#### Multiple Patient Prediction

```python
print("="*70)
print("Multiple Patient Prediction")
print("="*70)

patients = pd.DataFrame([

{
"Age":22,
"SystolicBP":120,
"DiastolicBP":80,
"BS":6.5,
"BodyTemp":98.2,
"HeartRate":75
},

{
"Age":33,
"SystolicBP":145,
"DiastolicBP":95,
"BS":8.4,
"BodyTemp":99.1,
"HeartRate":90
},

{
"Age":41,
"SystolicBP":160,
"DiastolicBP":100,
"BS":12.1,
"BodyTemp":101.0,
"HeartRate":102
}

])

patients_processed = preprocess_input(patients)

patients["Predicted Risk"] = model.predict(patients_processed)

display(patients)
```

#### Prediction Summary

```python
print("="*70)
print("Prediction Summary")
print("="*70)

print("Total Patients :", len(patients))

print()

print(
patients["Predicted Risk"].value_counts()
)
```
