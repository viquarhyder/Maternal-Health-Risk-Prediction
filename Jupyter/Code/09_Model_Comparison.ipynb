# Maternal Health Risk Prediction

## Model Comparison

#### Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Model Comparison Started")
```

#### Model Performance Summary

```python
comparison = pd.DataFrame({

    "Model":[

        "Random Forest",
        "XGBoost",
        "Support Vector Classifier",
        "Artificial Neural Network",
        "K-Nearest Neighbors"

    ],

    "Accuracy":[

        85.71,
        84.23,
        83.74,
        88.91,
        86.20

    ],

    "Precision":[

        85.60,
        84.10,
        83.40,
        88.70,
        86.10

    ],

    "Recall":[

        85.70,
        84.20,
        83.70,
        88.90,
        86.20

    ],

    "F1 Score":[

        85.60,
        84.10,
        83.50,
        88.80,
        86.10

    ]

})

comparison
```

#### Best Performing Model

```python
best_model = comparison.loc[

    comparison["Accuracy"].idxmax()

]

best_model
```

#### Accuracy Comparison

```python
plt.figure(figsize=(9,5))

sns.barplot(

    data=comparison,

    x="Model",

    y="Accuracy"

)

plt.title("Model Accuracy Comparison")

plt.xticks(rotation=15)

plt.grid(alpha=.3)

plt.show()
```

#### Precision Comparison

```python
plt.figure(figsize=(9,5))

sns.barplot(

    data=comparison,

    x="Model",

    y="Precision"

)

plt.title("Model Precision Comparison")

plt.xticks(rotation=15)

plt.grid(alpha=.3)

plt.show()
```

#### Recall Comparison

```python
plt.figure(figsize=(9,5))

sns.barplot(

    data=comparison,

    x="Model",

    y="Recall"

)

plt.title("Model Recall Comparison")

plt.xticks(rotation=15)

plt.grid(alpha=.3)

plt.show()
```

#### F1 Score Comparison

```python
plt.figure(figsize=(9,5))

sns.barplot(

    data=comparison,

    x="Model",

    y="F1 Score"

)

plt.title("Model F1 Score Comparison")

plt.xticks(rotation=15)

plt.grid(alpha=.3)

plt.show()
```

#### Overall Performance Heatmap

```python
plt.figure(figsize=(8,5))

sns.heatmap(

    comparison.set_index("Model"),

    annot=True,

    cmap="YlGnBu",

    fmt=".2f"

)

plt.title("Overall Model Performance")

plt.show()
```

#### Final Conclusion

```python
print("="*70)

print("BEST MODEL")

print("="*70)

print()

print(best_model)

print()

print("Selected Model :", best_model["Model"])

print()

print("Accuracy :", best_model["Accuracy"],"%")
```

```python
comparison.plot(

    x="Model",

    y=["Accuracy","Precision","Recall","F1 Score"],

    kind="bar",

    figsize=(10,6)

)

plt.title("Overall Model Comparison")

plt.ylabel("Performance (%)")

plt.grid(alpha=.3)

plt.xticks(rotation=15)

plt.show()
```
